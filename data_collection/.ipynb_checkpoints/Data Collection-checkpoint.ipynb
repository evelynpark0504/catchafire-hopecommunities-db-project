```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.firefox.options import Options
import pickle

# Set Firefox headless mode
options = Options()
options.headless = True
driver = webdriver.Firefox(executable_path='/usr/local/bin/geckodriver', options=options)

property_database = []

# Loop through the first 18 pages
for page in range(1, 18):
    url = f"https://www.apartments.com/denver-co/{page}/"
    driver.get(url)

    # Allow time for JavaScript to load
    time.sleep(5)

    # Extract the apartment listings
    apartments = driver.find_elements(By.CLASS_NAME, 'placard')

    # Loop through the results
    for apt in apartments:
        try:
            # Get the link to the apartment website
            website = apt.find_element(By.TAG_NAME, 'a').get_attribute('href')
            
            try:
                name = apt.find_element(By.CLASS_NAME, 'property-title').text
            except:
                name = "N/A"
                
            try:
                address = apt.find_element(By.CLASS_NAME, 'property-address').text
            except:
                address = "N/A"

            try:
                price = apt.find_element(By.CLASS_NAME, 'property-pricing').text
            except:
                price = "N/A"
            
            property_dict = {}
            property_dict['website'] = website
            property_dict['name'] = name
            property_dict['address'] = address
            
            property_database.append(property_dict)

            # Print the details
            print(f"Website: {website}")
            print(f"Apartment Name: {name}")
            print(f"Address: {address}")
            print(f"Price: {price}")
            print('-' * 50)

        except Exception as e:
            print("Error extracting data for this listing:", e)
            with open('property_db.pkl', 'wb') as f:
                pickle.dump(property_database, f)
            

# Close the browser window
driver.quit()
```

```python
url = "https://www.apartments.com/denver-co/"
driver.get(url)
time.sleep(5)

# Get the total number of pages (assuming the pagination element is available)
try:
    pagination = driver.find_element(By.CLASS_NAME, 'paging')
    total_pages = int(pagination.find_elements(By.TAG_NAME, 'a')[-2].text)  # Grabs the second-to-last link which usually contains the last page number
except Exception as e:
    print("Error extracting total pages:", e)
    total_pages = 1  # Default to 1 page if unable to find pagination

for page in range(1, total_pages + 1):
    if page > 1:
        # If not on the first page, navigate to the next page
        url = f"https://www.apartments.com/denver-co/{page}/"
        driver.get(url)
        time.sleep(5)

    # Extract the apartment listings
    apartments = driver.find_elements(By.CLASS_NAME, 'placard')

    # Loop through the results
    for apt in apartments:
        try:
            # Get the link to the apartment website
            website = apt.find_element(By.TAG_NAME, 'a').get_attribute('href')

            try:
                name = apt.find_element(By.CLASS_NAME, 'property-title').text
            except:
                name = "N/A"

            try:
                address = apt.find_element(By.CLASS_NAME, 'property-address').text
            except:
                address = "N/A"

            try:
                price = apt.find_element(By.CLASS_NAME, 'property-pricing').text
            except:
                price = "N/A"

            # Store the data in a dictionary
            property_dict = {
                'website': website,
                'name': name,
                'address': address,
                'price': price
            }
            
            # Append the data to the database
            property_database.append(property_dict)

            # Print the details
            print(f"Website: {website}")
            print(f"Apartment Name: {name}")
            print(f"Address: {address}")
            print(f"Price: {price}")
            print('-' * 50)

        except Exception as e:
            print("Error extracting data for this listing:", e)
            with open('property_db.pkl', 'wb') as f:
                pickle.dump(property_database, f)

    # Save progress after each page
    with open('property_db.pkl', 'wb') as f:
        pickle.dump(property_database, f)

# Close the webdriver after completion
driver.quit()
```
