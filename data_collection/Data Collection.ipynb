```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.firefox.options import Options
import pickle
import re

# Set Firefox headless mode
options = Options()
options.headless = True
driver = webdriver.Firefox(executable_path='/usr/local/bin/geckodriver', options=options)

property_database = []

url = "https://www.apartments.com/denver-co/"
driver.get(url)
time.sleep(5)

# Get the total number of pages from the "Showing X of Y Results - Page N of M" text
try:
    # Locate the element containing the results summary
    results_summary = driver.find_element(By.CLASS_NAME, 'pageRange').text
    match = re.search(r'Page \d+ of (\d+)', results_summary)
    if match:
        total_pages = int(match.group(1))  # Get the total number of pages
    else:
        total_pages = 1  # Default to 1 if the total pages cannot be found
except Exception as e:
    print("Error extracting total pages:", e)
    total_pages = 1  # Default to 1 page if unable to find the results summary


for page in range(1, total_pages):
    if page > 1:
        url = f"https://www.apartments.com/denver-co/{page}/"
        driver.get(url)
        time.sleep(5)  # Allow time for the page to load

    apartments = driver.find_elements(By.CLASS_NAME, 'placard')

    # Loop through the results
    for apt in apartments:
        try:
            # Get the link to the apartment website
            website = apt.find_element(By.TAG_NAME, 'a').get_attribute('href')
            if website == "javascript:void(0);": continue;
            else:
                try:
                    name = apt.find_element(By.CLASS_NAME, 'property-title').text
                except:
                    name = "N/A"

                try:
                    address = apt.find_element(By.CLASS_NAME, 'property-address').text
                except:
                    address = "N/A"

                try:
                    price = apt.find_element(By.CLASS_NAME, 'property-pricing').text
                except:
                    price = "N/A"

                property_dict = {
                    'website': website,
                    'name': name,
                    'address': address,
                    'price': price
                }

                # Append the data to the database
                property_database.append(property_dict)

                # Print the details
                print(f"Website: {website}")
                print(f"Apartment Name: {name}")
                print(f"Address: {address}")
                print(f"Price: {price}")
                print('-' * 50)

        except Exception as e:
            print("Error extracting data for this listing:", e)
            with open('property_db.pkl', 'wb') as f:
                pickle.dump(property_database, f)

    # Save progress after each page
    with open('property_db.pkl', 'wb') as f:
        pickle.dump(property_database, f)

# Close the webdriver after completion
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import time

# Set Firefox headless mode
options = Options()
options.headless = True
driver = webdriver.Firefox(executable_path='/usr/local/bin/geckodriver', options=options)
driver.set_page_load_timeout(60)

property_details = []

for prop in property_database:
    url = prop['website']
    
    try:
        driver.get(url)

        try:
            grid_items = driver.find_elements(By.CSS_SELECTOR, '.pricingGridItem.multiFamily.hasUnitGrid')


            for grid_item in grid_items:
                amenity_dict = {}
                amenities_section = grid_item.find_element(By.CLASS_NAME, 'allAmenities')
                categories = amenities_section.find_elements(By.CLASS_NAME, 'topAmenity')

                for category in categories:
                    category_name = category.get_attribute('innerText')
                    parent_li = category.find_element(By.XPATH, './..')
                    amenities = parent_li.find_elements(By.CLASS_NAME, 'amenity')
                    amenity_list = []

                    for amenity in amenities:
                        amenity_text = amenity.get_attribute('innerText')
                        amenity_list.append(amenity_text)

                    amenity_dict[category_name] = amenity_list

                unit_type = grid_item.find_element(By.CLASS_NAME, 'detailsTextWrapper')
                unit_details = [span.text for span in unit_type.find_elements(By.TAG_NAME, 'span')]

                if len(unit_details) >= 3:
                    bed = unit_details[0]
                    bath = unit_details[1]
                else:
                    bed = "N/A"
                    bath = "N/A"

                units = grid_item.find_elements(By.CLASS_NAME, 'unitContainer')

                for unit in units:
                    try:
                        unitbtn = unit.find_element(By.CLASS_NAME, 'unitBtn').text
                    except:
                        unitbtn = "N/A"

                    try:
                        price = unit.find_element(By.CLASS_NAME, 'pricingColumn').text
                    except:
                        price = "N/A"

                    try:
                        sqft = unit.find_element(By.CLASS_NAME, 'sqftColumn').text
                    except:
                        sqft = "N/A"

                    if unit!="" and price!="":
                        property_dict = {}
                        property_dict['website'] = url
                        property_dict['address'] = prop['address']
                        property_dict['unit_number'] = unitbtn
                        property_dict['bed'] = bed
                        property_dict['bath'] = bath
                        property_dict['amenity'] = amenity_dict
                        property_dict['price'] = price
                        property_dict['sqft'] = sqft

                        property_details.append(property_dict)
                        print(f"in progress... {url}, unit: {unitbtn}, bed: {bed}, price: {price}")
                        time.sleep(5)

        except Exception as e:
            print("Error extracting unit prices or square footage:", e)
            with open('property_details.pkl', 'wb') as f:
                pickle.dump(property_details, f)
    except TimeoutException:
        print("Page took too long to load.")

# Close the browser after scraping
driver.quit()
```

```python
import re

for detail in property_details:
    price_int = re.sub(r"[^\d]", "", detail['price'])
    sqft_int = re.sub(r"[^\d]", "", detail['sqft'])
    print(f"price: {price_int}, sqft: {sqft_int}")
    
    if price_int!="":
        detail['price'] = int(price_int)
    else:
        detail['price'] = -1
        
    if sqft_int!="":
        detail['sqft'] = int(sqft_int)
    else:
        detail['sqft'] = -1
```
