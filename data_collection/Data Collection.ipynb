```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.firefox.options import Options
import pickle
import re

# Set Firefox headless mode
options = Options()
options.headless = True
driver = webdriver.Firefox(executable_path='/usr/local/bin/geckodriver', options=options)

property_database = []

url = "https://www.apartments.com/denver-co/"
driver.get(url)
time.sleep(5)

# Get the total number of pages from the "Showing X of Y Results - Page N of M" text
try:
    # Locate the element containing the results summary
    results_summary = driver.find_element(By.CLASS_NAME, 'pageRange').text
    match = re.search(r'Page \d+ of (\d+)', results_summary)
    if match:
        total_pages = int(match.group(1))  # Get the total number of pages
    else:
        total_pages = 1  # Default to 1 if the total pages cannot be found
except Exception as e:
    print("Error extracting total pages:", e)
    total_pages = 1  # Default to 1 page if unable to find the results summary


for page in range(1, total_pages):
    if page > 1:
        url = f"https://www.apartments.com/denver-co/{page}/"
        driver.get(url)
        time.sleep(5)  # Allow time for the page to load

    apartments = driver.find_elements(By.CLASS_NAME, 'placard')

    # Loop through the results
    for apt in apartments:
        try:
            # Get the link to the apartment website
            website = apt.find_element(By.TAG_NAME, 'a').get_attribute('href')

            try:
                name = apt.find_element(By.CLASS_NAME, 'property-title').text
            except:
                name = "N/A"

            try:
                address = apt.find_element(By.CLASS_NAME, 'property-address').text
            except:
                address = "N/A"

            try:
                price = apt.find_element(By.CLASS_NAME, 'property-pricing').text
            except:
                price = "N/A"

            # Store the data in a dictionary
            property_dict = {
                'website': website,
                'name': name,
                'address': address,
                'price': price
            }

            # Append the data to the database
            property_database.append(property_dict)

            # Print the details
            print(f"Website: {website}")
            print(f"Apartment Name: {name}")
            print(f"Address: {address}")
            print(f"Price: {price}")
            print('-' * 50)

        except Exception as e:
            print("Error extracting data for this listing:", e)
            with open('property_db.pkl', 'wb') as f:
                pickle.dump(property_database, f)

    # Save progress after each page
    with open('property_db.pkl', 'wb') as f:
        pickle.dump(property_database, f)

# Close the webdriver after completion
driver.quit()
```
